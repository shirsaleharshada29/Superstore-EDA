```python
#1. Data Loading

# Load the dataset

import pandas as pd

data=pd.read_excel(r"C:\Users\Shirs\Downloads\Sample - Superstore.xlsx")
print(data)
```

```python
print(data.info)
```

```python
# 2. Data Cleaning

# Check for missing values

print(data.isnull().sum())



# percent
a=(data.isnull().sum() / len(data)) * 100
print(a)
```

```python
# Handle missing values (mean, median, mode, or drop)(optional)

# Replace missing postal code values with mode (most appropriate)
data['Postal Code'] = data['Postal Code'].fillna(data['Postal Code'].mode()[0])

data['Postal Code'] = data['Postal Code'].astype(int)

print(data['Postal Code'])   
       
```

```python
# Check data types and convert if necessary

data.dtypes
```

```python
#  Check for duplicates

data.duplicated().sum()


```

```python
# 3. Visualization

# Use the following plots to visualize key aspects of the data:

import matplotlib.pyplot as plt
import seaborn as sns

# Bar Chart: Use for Category

plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Category', palette='viridis')
plt.title("Count of Categories")
plt.xlabel("Category")
plt.ylabel("Count")

plt.savefig("category_count.png", dpi=300, bbox_inches="tight")
plt.show()

```

```python
# Sub-Category

plt.figure(figsize=(12, 5))
sns.countplot(data=data, x='Sub-Category', palette='magma')
plt.title("Count of Sub-Categories")
plt.xlabel("Sub-Category")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.savefig("sub_category_count.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Segment

plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Segment', palette='cool')
plt.title("Count of Segments")
plt.xlabel("Segment")
plt.ylabel("Count")

plt.savefig("segment_count.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Region

plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Region', palette='Set2')
plt.title("Count of Regions")
plt.xlabel("Region")
plt.ylabel("Count")

plt.savefig("region_count.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# State

plt.figure(figsize=(12, 6))
top_states = data['State'].value_counts().head(10)

sns.barplot(x=top_states.index, y=top_states.values, palette='plasma')
plt.title("Top 10 States by Number of Records")
plt.xlabel("State")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.savefig("top_10_states.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Scatter Plot: Use to compare Sales vs Profit

plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Sales', y='Profit', hue='Category', alpha=0.6)

plt.title("Scatter Plot: Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.legend(title="Category")

plt.savefig("sales_vs_profit_scatter.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#  Discount vs Profit

plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Discount', y='Profit', hue='Category', alpha=0.6)

plt.title("Scatter Plot: Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.legend(title="Category")

plt.savefig("discount_vs_profit_scatter.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Histogram: Use for numeric distributions such as Sales, Profit, Quantity, Discount

plt.figure(figsize=(9, 7))

plt.subplot(2, 2, 1)
plt.hist(data['Sales'], bins=30, color='skyblue')
plt.title("Sales")

plt.subplot(2, 2, 2)
plt.hist(data['Profit'], bins=30, color='salmon')
plt.title("Profit")

plt.subplot(2, 2, 3)
plt.hist(data['Quantity'], bins=30, color='lightgreen')
plt.title("Quantity")

plt.subplot(2, 2, 4)
plt.hist(data['Discount'], bins=30, color='plum')
plt.title("Discount")

plt.tight_layout()

plt.savefig("histograms_sales_profit_quantity_discount.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
# Heatmap: Use to visualize correlations between numeric variables such as Sales, Profit, Discount, and Quantity

corr_data = data[['Sales', 'Profit', 'Discount', 'Quantity']]

# Compute correlation matrix
corr_matrix = corr_data.corr()

# Plot heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")

plt.savefig("correlation_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# 4. Outlier Detection

# Use boxplots to identify outliers in Sales, Profit, and Discount

plt.figure(figsize=(9, 4))

# Boxplot for Sales
plt.subplot(1, 3, 1)
plt.boxplot(data['Sales'], patch_artist=True,
            boxprops=dict(facecolor='skyblue'))
plt.title("Sales")

# Boxplot for Profit
plt.subplot(1, 3, 2)
plt.boxplot(data['Profit'], patch_artist=True,
            boxprops=dict(facecolor='lightgreen'))
plt.title("Profit")

# Boxplot for Discount
plt.subplot(1, 3, 3)
plt.boxplot(data['Discount'], patch_artist=True,
            boxprops=dict(facecolor='salmon'))
plt.title("Discount")

plt.tight_layout()

plt.savefig("boxplots_sales_profit_discount.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```
